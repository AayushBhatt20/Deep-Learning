# TASK-1 : Activation Functions Implementation 

Activation functions are part of the neural network which determines if a neuron fires or not. 

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Binary Step Activation Function

Binary step function returns value either 0 or 1.
  + It returns '0' if the input is the less then zero
  + It returns '1' if the input is greater than zero

```python
def binaryStep(x):
    return np.heaviside(x,1)
x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation Function :binaryStep')
plt.show()
```

# 2. Linear Activation Function

In linear Activation Function
+ Output = Input

```python
def linear(x):
    return x

x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function :Linear')
plt.show()
```

# 3. Sigmoid Activation Function

Sigmoid function returns the value beteen 0 and 1. For activation function in deep learning network, Sigmoid function is considered not good since near the boundaries the network doesn't learn quickly. This is because gradient is almost zero near the boundaries.

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

# 4. Tanh Activation Function

Tanh is another nonlinear activation function. Tanh outputs between -1 and 1. Tanh also suffers from gradient problem near the boundaries just as Sigmoid activation function does.

```python
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```

# 5.  RELU Activation Function



RELU is more well known activation function which is used in the deep learning networks. RELU is less computational expensive than the other non linear activation functions.

  + RELU returns 0 if the x (input) is less than 0
  + RELU returns x if the x (input) is greater than 0



```python
def RELU(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1

x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :RELU')
plt.show()
```

# 6. Softmax Activation Function

Softmax turns logits, the numeric output of the last linear layer of a multi-class classification neural network into probabilities.

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)

x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.show()
```

# 7. Leaky RELU Activation function

Leaky ReLU function is nothing but an improved version of the ReLU function. As we saw that for the ReLU function, the gradient is 0 for x<0, which would deactivate the neurons in that region.

Leaky ReLU is defined to address this problem. Instead of defining the Relu function as 0 for negative values of x, we define it as an extremely small linear component of x

```python
#Here the small linear component x value is 0.05.

def lrelu(x):
    x2=[]
    for i in x:
        if i<0:
            x2.append(0.05*i) 
        else:
            x2.append(i)
    return x2

x = np.linspace(-10, 10)
plt.plot(x,lrelu(x))
plt.axis('tight')
plt.title('Activation Function :Leaky RELU Fumction')
plt.show()
```

# 8. Exponential Linear Unit

Exponential Linear Unit or ELU for short is also a variant of Rectiufied Linear Unit (ReLU) that modifies the slope of the negative part of the function. Unlike the leaky relu and parametric ReLU functions, instead of a straight line, ELU uses a log curve for defning the negatice values. 

```python
def elu(arr, alpha):
    a = []
    for x in arr:
        if x >= 0:
            a.append(x)
        else:
            a.append(alpha * (np.exp(x)-1))
    return a

X = np.linspace(-10, 10)
elu1 = elu(X, 1.0)
plt.plot(X, elu1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ELU')
plt.show()
```
